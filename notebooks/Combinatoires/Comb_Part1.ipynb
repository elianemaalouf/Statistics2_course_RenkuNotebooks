# Exercice 4

Simplifier les expressions suivantes:

1) $\frac{100!}{98!}$ 

2) $1000 . 999!$ 

3) $\frac{(n-1)!}{(n-3)!}$

1) $\frac{100!}{98!} = \frac{100 . 99 . 98!}{98!} = 100 . 99 = 9900$. On simplifie $98!$ du num√©rateur et du d√©nominateur. 

2) $1000. 999! = 1000! \approx \sqrt{2000\pi}.\large(\frac{1000}{e}\large)^{1000}$ C'est un tr√®s grand chiffre, il doit √™tre approxim√© en pratique.

3) $\frac{(n-1)!}{(n-3)!} = \frac{(n-1)(n-2)(n-3)!}{(n-3)!} = (n-1)(n-2) = n^2 -2n -n + 2 = n^2 -3n + 2$
